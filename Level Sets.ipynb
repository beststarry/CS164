## Level Sets

Here, we attempt to plot level sets of the Cobb-Douglas function, which is of the form $$f(x,y) = 100 * x^{3/4} * y^{1/4} $$

furthermore, we are given the constraint $$ 200x + 250y = 50000 $$

In order to find the maximum, first we do visualizations to find the level sets of the production function.


```python
import numpy as np
import matplotlib.pyplot as plt

delta = 1
x = np.arange(0, 251)
y = np.arange(0, 201)

X, Y = np.meshgrid(x, y)
Z = 100 * X**(3/4) * Y**(1/4)

fig, ax = plt.subplots(figsize = (10,6))
CS = ax.contour(X, Y, Z)

plt.plot(x, (50000 - 200*x)/250 )
ax.clabel(CS)
ax.set_title('Contour Plots of the Cobb - Douglas Function')

plt.show()
```

```python
import sympy as sp
from sympy.tensor.array import derive_by_array

x, y = sp.symbols('x, y', real = True)
f = 100 * x**(3/4) * y*(1/4)
g = 200 * x + 250 * y - 50000

print(derive_by_array(f,(x,y)))
print(derive_by_array(g,(x,y)))
```
